```python
import pandas as pd

df = pd.read_excel('Depositantes_MAB.xlsx')

# Create a new index column and drop the default numerical index
df = df.reset_index()
df
```

```python
list_of_strings = []
for my_string in df['Rótulos de Linha']:
    split_result = my_string.split(',')
    #print(split_result[0])
    list_of_strings.append(split_result[0])
```

```python
df_separate = pd.DataFrame(list_of_strings, columns=['Empresas'])
df_separate = df_separate.reset_index()
df_separate
```

```python
import re
from unicodedata import normalize

# Function to eliminate unwanted characters
def __normalize_str(_str):
    return re.sub(
        r"[,;{}()\n\t=-]",
        "",
        normalize("NFKD", _str)
        .encode("ASCII", "ignore")
        .decode("ASCII")
        .replace(" ", "_")
        .replace("-", "_")
        .upper(),
    )

# Apply the clean_text function to the 'text' column
df_separate['cleaned_text'] = df_separate['Empresas'].apply(__normalize_str)
```

```python
df_separate
```

```python
# Join
# Merge dataframes on the 'ID' column
merged_df = pd.merge(df, df_separate, on='index', how='left')  # You can use 'left', 'right', or 'outer' as well

# Display the resulting merged dataframe
merged_df
```

```python
merged_df[['index', 'Rótulos de Linha', 'cleaned_text', 'Contagem de Publication Number']].to_csv("Depositantes_MAB_Limpos.csv", index=False)
```
